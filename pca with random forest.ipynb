```python
#import library 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics



#load data
data = pd.read_csv('Dataset_VPN.csv',delimiter=',')
Null=data.isnull().any().sum()
print("Hasil Pengecekkan Data kosong pada dataset :",Null)
print('\n')

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data['labels'] = labelencoder.fit_transform(data['class1'])
dataset= data.drop(['Unnamed: 0','class1'],axis = 1)
print('Dataset Sebelum direduksi')
display(dataset)
print ("Komposisi Data VPN dan Non VPN")
print (dataset.labels.value_counts())
print('\n')

#Reduction dataset using PCA
features = dataset.iloc[:, 0:22].values
labels = dataset.iloc[:, 23].values
scaler = MinMaxScaler()
featuress = scaler.fit_transform(features)
pca = PCA(n_components=2)
pca.fit(features)
X_pca = pca.transform(features)
reduced_data = pd.DataFrame(X_pca)
reduced_data = reduced_data.add_prefix('feature_')
reduced_data['labels']=labels
print('Dataset Sesudah direduksi')
display(reduced_data.head())
#viz reduced dataset
print("\n")
print("Visualisasi Reduce data using PCA")
sns.scatterplot(reduced_data['feature_0'],reduced_data['feature_1'],hue=reduced_data['labels'])
plt.title('Scatter-plot')
plt.show()

#Classification using Random Forest
x = reduced_data.iloc[:, 0:1].values
y = reduced_data.iloc[:, 2].values
X = scaler.fit_transform(x)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

model = RandomForestClassifier(n_estimators=5)
model.fit(X_train,y_train)
model_pred = model.predict(X_test)

from sklearn import metrics
print("=== Accuracy ====")
print(metrics.accuracy_score(y_test, model_pred))
print('\n')
from sklearn.metrics import classification_report, confusion_matrix
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, model_pred))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, model_pred))
print('\n')


def plot_roc_curve(fpr, tpr):
    plt.plot(fpr, tpr, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()

from sklearn.metrics import roc_curve  
fpr, tpr, thresholds = roc_curve(y_test, model_pred)
plot_roc_curve(fpr, tpr)
```
